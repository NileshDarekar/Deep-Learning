## Lithology Classification, generated data.

Features: Porosity and Perm.

Label: Limestone or sandstone


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# data sandstone

phi_s = np.abs(np.random.normal(0.25,0.01,300))
swc_s = np.abs(np.random.normal(0.1,0.01,300))
k_s = 8.58102*(phi_s**(4.4))/(swc_s**2) ## Timur equation
plt.scatter(phi_s,k_s)
plt.xlabel('porosity')
plt.ylabel('perm')
```

```python
#Data
#Limestone
phi_l = np.abs(np.random.normal(0.20,0.01,300))
swc_l = np.abs(np.random.normal(0.1,0.01,300))
k_l = 8.58102*(phi_l**(4.4))/(swc_l**2)
plt.scatter(phi_l,k_l)
plt.scatter(phi_s,k_s)
plt.xlabel('porosity')
plt.ylabel('perm(D)')
```

```python
#DataFrame Sandstone
dfS = pd.DataFrame({'Porosity':phi_s , 'Swc':swc_s, 'Permeability(D)':k_s})
dfS['label'] = 1
label_s = np.array(dfS['label'])
dfS['lithology'] = "Sandstone"
lithologyS = np.array(dfS['lithology'])
dfS
```

```python
#DataFrame Limestone
dfL = pd.DataFrame({'Porosity':phi_l , 'Swc':swc_l, 'Permeability(D)':k_l})
dfL['label'] = 0
label_l = np.array(dfL['label'])
dfL['lithology'] = "Limestone"
lithologyL = np.array(dfL['lithology'])
dfL
```

```python
df_final = dfL.append(dfS,ignore_index=True)
```

## Modelling

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
def create_baseline():
    #Create Model
    model = Sequential()
    model.add(Dense(60,input_dim = 2,activation = 'sigmoid'))
    model.add(Dense(30,activation = 'sigmoid'))
    model.add(Dense(1,activation = 'sigmoid'))
    
    #Compile Model
    model.compile(loss = 'binary_crossentropy',optimizer = 'sgd',metrics = ['accuracy'])
    return model
```

```python
model = create_baseline()
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,validation_split=0.2,epochs = 100,batch_size=32)
```

One Hot Encoding

```python
def one_hot(y):
    n_values = np.max(y) + 1
    y_new = np.eye(n_values)[y[:,0]]
    return y_new
```

```python
y = one_hot(y)
```

```python
y
```

```python
x_train , x_test, y_train, y_test = train_test_split(X, y, test_size =.1, random_state = 27)
```

```python
y_train[0]
```

```python
from tensorflow.keras.optimizers import Adam
```

## TensorFlow has a SparseCategoricalCrossentropy and a CategoricalCrossentropy loss function. The first expect your labels to be provided as integers, where the latter expects one-hot encodings.

```python
def create_baseline_onehotencoded():
    # create model
    model = Sequential()
    model.add(Dense(60, input_dim=2, activation='sigmoid'))
    model.add(Dense(30, activation='sigmoid'))
    model.add(Dense(2, activation='sigmoid'))
    # Compile model
    model.compile(loss='categorical_crossentropy', optimizer=Adam(lr=0.001), metrics=['accuracy'])
    return model
```

```python
model = create_baseline_onehotencoded()
```

```python
model.summary()
```

```python
history = model.fit(x_train,y_train,epochs = 100,batch_size=32,validation_split=0.2)
```

```python
history.history.keys()
```

```python
df = pd.DataFrame()
for i in history.history.keys():
    df[i] = history.history[i]
```

```python
df
```

```python
df.plot(subplots = True,figsize = (10,7))
```

```python
plt.figure(figsize = (10,5))
plt.plot(df['loss'],label = 'Training Loss')
plt.plot(df['val_loss'],label = 'Val Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Losses')
plt.legend()
plt.grid(True)
```

```python
plt.figure(figsize = (10,5))
plt.plot(df['accuracy'],label = 'Training Accuracy')
plt.plot(df['val_accuracy'],label = 'Val Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Losses')
plt.legend()
plt.grid(True)
```
